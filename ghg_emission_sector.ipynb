## Global greenhouse gas emissions: by sector

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# loading data
gm_main_sector = pd.read_csv('data/GHG-Emissions-by-sector.csv')
gm_sub_sector = pd.read_csv('data/GHG-Emissions-by-sub_sector.csv')
gm_sub_sector_breakdown = pd.read_csv('data/GHG-Emissions-by-sub_sector(Furthure breakdown).csv')
```

```python
# Gave shorter column names 
def rename_df_col(df):
    df = df.rename(columns = {'Share of global greenhouse gas emissions (%)': 'Share'})
    return df

gm_main_sector = rename_df_col(gm_main_sector)
gm_sub_sector = rename_df_col(gm_sub_sector)
gm_sub_sector_breakdown = rename_df_col(gm_sub_sector_breakdown)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.set_style('dark')
sns.barplot(x=gm_main_sector['Sector'], y=gm_main_sector['Share'], palette='muted')
ax.set_title('GHG Emission By Main Sectors')
```

```python
fig, ax = plt.subplots(figsize=(12,6))
sns.set_style('dark')
sns.barplot(x=gm_sub_sector['Sub-sector'], y=gm_sub_sector['Share'], hue=gm_sub_sector['Sector'], palette='muted')
ax.set_xticklabels(gm_sub_sector['Sub-sector'], rotation=90)
```
